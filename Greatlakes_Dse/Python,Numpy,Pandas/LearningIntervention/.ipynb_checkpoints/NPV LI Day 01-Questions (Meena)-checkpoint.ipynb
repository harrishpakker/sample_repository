<p style="background-color:#b5e7a0;font-size:18px;color:#4040a1" size="4"><b> Exercise 12: Read the covid_data csv file</b> </p>

```python
covid_data = pd.read_csv('Covid_data.csv')
covid_data
```

<p style="background-color:#b5e7a0;font-size:18px;color:#4040a1" size="4"><b> Exercise 13: Drop the S.No. column </b> </p>

```python
covid_data.drop('S. No.',axis= 'columns',inplace=True)
```

<p style="background-color:#b5e7a0;font-size:18px;color:#4040a1" size="4"><b> Exercise 14: Remove the blank spaces present in the column names </b> </p>

```python
covid_data.columns = covid_data.columns.str.replace(' ','')
```

```python
covid_data.columns
```

<p style="background-color:#b5e7a0;font-size:18px;color:#4040a1" size="4"><b> Exercise 15: Remove the , symbol present in the records </b> </p>

```python
covid_data['Recovered'] = covid_data['Recovered'].str.replace(',','')
covid_data['ActiveCases'] = covid_data['ActiveCases'].str.replace(',','')
covid_data['deceased'] = covid_data['deceased'].str.replace(',','')
```

<p style="background-color:#b5e7a0;font-size:18px;color:#4040a1" size="4"><b> Exercise 16: Replace the string '-' present in the deceased column with 0  </b> </p>

```python
covid_data['Recovered'] = covid_data['Recovered'].str.replace('-','0')
covid_data['ActiveCases'] = covid_data['ActiveCases'].str.replace('-','0')
covid_data['deceased'] = covid_data['deceased'].str.replace('-','0')
```

<p style="background-color:#b5e7a0;font-size:18px;color:#4040a1" size="4"><b> Exercise 17: Check the data information, convert the column with values in numbers to int data type </b> </p>

```python
covid_data['Recovered'] = covid_data['Recovered'].astype(int)
covid_data['ActiveCases'] = covid_data['ActiveCases'].astype(int)
covid_data['deceased'] = covid_data['deceased'].astype(int)
```

<p style="background-color:#b5e7a0;font-size:18px;color:#4040a1" size="4"><b> Exercise 21: Load the emp1, emp2 and dept datasets </b> </p>

```python
emp1 = pd.read_csv('Emp_data_1.csv')
emp2 = pd.read_csv('Emp_data_2.csv')
dept = pd.read_csv('Dept_data.csv')
```

```python
dept.head()
```

<p style="background-color:#b5e7a0;font-size:18px;color:#4040a1" size="4"><b> Exercise 22: Concatenate the employee1 and 2 datasets </b> </p>

```python
employee_data = pd.concat([emp1,emp2],ignore_index=True,axis=0)
```

```python
employee_data.rename({'department_id':'Dept_id'},axis =1,inplace=True)
employee_data
```

<p style="background-color:#b5e7a0;font-size:18px;color:#4040a1" size="4"><b> Exercise 23: Merge the employee and department table, on column department id. Ensure that none of the record is missed out </b> </p>

```python
outer_join = pd.merge(employee_data,dept,how='outer',on = 'Dept_id')
```

```python
outer_join
```

<p style="background-color:#b5e7a0;font-size:18px;color:#4040a1" size="4"><b> Exercise 23: Join the dept and emp table on dept id, display only the matching records </b> </p>

```python
inner_join = pd.merge(employee_data,dept,how='inner',on = 'Dept_id')
```

```python
inner_join
```

<p style="background-color:#b5e7a0;font-size:18px;color:#4040a1" size="4"><b> Exercise 24: Display all the records in employee table and the matching records in department table </b> </p>

```python
left_join = pd.merge(employee_data,dept,how='left',on = 'Dept_id')
left_join
```

<p style="background-color:#b5e7a0;font-size:18px;color:#4040a1" size="4"><b> Exercise 25: Display all the records in department table and the matching records in employee table </b> </p>

```python
right_join = pd.merge(employee_data,dept,how='right',on = 'Dept_id')
right_join
```
