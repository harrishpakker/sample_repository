```python
# Syntax:

# df['colname'].replace({oldvalue: newvalue},inplace = True/False) # To replace in the single column
# df.replace({oldvalue: newvalue},inplace = True/False) # To replace in the entire dataframe.
```

```python
names = ['Ram', 'John', 'Abdul','Siva', 'Adam', 'Rahman']
maths =  [45, 40, 60, 80, 90, 75]
english =[60,75,90,65,80,60]
di={'Names':names,'Maths':maths,'English':english}
scores_df=pd.DataFrame(di)
scores_df
```

```python
# Replace Mark 60 -> 65 in the maths column
```

```python
scores_df['Maths'].replace({60:65})  # display the change and not modifying the orginal df.
```

```python
scores_df['Maths']= scores_df['Maths'].replace({60:65})  # Making the change in orignal df
scores_df
```

```python
scores_df['Maths'].replace({45:46},inplace=True)
scores_df
```

```python
scores_df.replace({80:83})   # it replaces all the entires in the dataframe
```

```python
scores_df.iloc[0,2] = 70  # replacing the each and every element 
scores_df

# Syntax:
# df.iloc[row,col]= newvalue
```

<span style="font-family: Arial; font-weight:bold;font-size:2.00em;color:#9F1F1B;">  Conditional Fromatting or Filtering

```python
# Syntax

# df[col]-condition  - boolean df
# df[df[col]-condition] - filter the df
```

```python
names = ['Ram', 'John', 'Abdul','Siva', 'Adam', 'Rahman']
maths =  [45, 40, 60, 80, 90, 75]
english =[60,75,90,65,80,60]
di={'Names':names,'Maths':maths,'English':english}
scores_df=pd.DataFrame(di)
scores_df
```

```python
# Maths greater than 70

scores_df['Maths']>70    # Return a boolean df
```

```python
scores_df[scores_df['Maths']>70]  # It filters the values whose condition is true
```

```python
maths_70=scores_df[scores_df['Maths']>70]   
maths_70['Names']
```

```python
scores_df[scores_df['Maths']>70]['Names']
```

```python
scores_df[scores_df['Maths']>70][['Names','Maths']]
```

```python
# Conditonal formatting using mutiple condition
```

```python
# Syntax:

# df[(cond1)logicalsymbol(cond2)]
# only logical symbol needs to be used
# each conditon needs to be seperated by brackets
```

```python
names = ['Ram', 'John', 'Abdul','Siva', 'Adam', 'Rahman']
maths =  [45, 40, 60, 80, 90, 75]
english =[60,75,90,65,80,60]
di={'Names':names,'Maths':maths,'English':english}
scores_df=pd.DataFrame(di)
scores_df
```

```python
scores_df[(scores_df['Maths']>70) & (scores_df['English']>60)]  # Logical and -&
```

```python
scores_df[(scores_df['Maths']>70) and (scores_df['English']>60)] # error
```

```python
scores_df[(scores_df['Maths']>70) | (scores_df['English']>60)] # Logical or - |
```

```python
scores_df[~(scores_df['Maths']>70)]  # Logical not - ~
```

```python
# sort_index

# Syntax:

# df.sort_index(axis=1/0, inplace=True/False)

# axis = 0 - row wise
# axis = 1 - column wise
```

```python
names = ['Ram', 'John', 'Abdul','Siva', 'Adam', 'Rahman']
maths =  [45, 40, 60, 80, 90, 75]
english =[60,75,90,65,80,60]
di={'Names':names,'Maths':maths,'English':english}
scores_df=pd.DataFrame(di,index=[1,5,4,3,2,0])
scores_df
```

```python
scores_df.sort_index(axis=1,inplace=True)
scores_df                                  # sorted in the alphabetical order in the column
```

```python
scores_df.sort_index(axis=0,inplace=True)  # sorted in the ascending order in the row
scores_df
```

<span style="font-family: Arial; font-weight:bold;font-size:2.00em;color:#9F1F1B;">  Merge

```python

Image('2.png')
```

```python
left_frame = pd.DataFrame({'key': range(5), 
                           'left_value': ['a', 'b', 'c', 'd', 'e']})
left_frame
```

```python
right_frame = pd.DataFrame({'key': range(2, 7), 
                           'right_value': ['f', 'g', 'h', 'i', 'j']})
right_frame
```

<span style="font-family: Arial; font-weight:bold;font-size:2.00em;color:#9F1F1B;">  Inner Join

```python
Image('2.1.png')
```

```python
# # Syntax:

# pd.merge(df1,df2,on='columnname on which merge happen', how='type of joins')
```

```python
pd.merge(left_frame, right_frame, on='key', how='inner')
```

<span style="font-family: Arial; font-weight:bold;font-size:2.00em;color:#9F1F1B;"> Outer Join

```python
Image('2.2.png')
```

```python
pd.merge(left_frame, right_frame, on='key', how='outer')
```

<span style="font-family: Arial; font-weight:bold;font-size:2.00em;color:#9F1F1B;"> Right Join

```python
Image('2.3.png')
```

```python
pd.merge(left_frame, right_frame, on='key', how='right')
```

<span style="font-family: Arial; font-weight:bold;font-size:2.00em;color:#9F1F1B;"> Left Join

```python
Image('2.4.png')
```

```python
pd.merge(left_frame, right_frame, on='key', how='left')
```

```python
# import the library to xml file
import xml.etree.ElementTree as ET 

# extract the xml file
tree = ET.parse("student_data.xml")
root = tree.getroot() 

# assign column names of the output DataFrame
df_col = ["Name", "Gender", "Marks"]

# create an empty list 
rows = []

for node in root: 
    name = node.attrib.get("name")
    gender = node.find("gender").text if node is not None else None
    marks = node.find("marks").text if node is not None else None
    
    # append each observation in the data to ‘rows’
    rows.append({"Name": name, "Gender": gender, 
                 "Marks": marks})

# create a DataFrame    
df_student = pd.DataFrame(rows, columns = df_col)

# print the DataFrame
df_student
```
