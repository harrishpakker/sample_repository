### Iterating Numpy Array

```python
# Iterating Array


ar1=np.arange(1,10).reshape(3,3)
print(ar1)
print('*'*25)
for i in ar1:   # row wise iteration happens i.e each row is taken as an element
    print(i)
    print('-'*50)
```

```python
ar1=np.arange(1,7).reshape(2,3)
print(ar1)
print('*'*25)
for i in ar1:
    print(i)
    print('-'*50)
```

```python
ar1=np.arange(1,10).reshape(3,3)
ar1
```

```python
ar1.T
```

```python
for i in ar1.T:   # Columnwise iteration
    print(i)
```

```python
# iterating element wise
```

```python
# function : np.nditer()
```

```python
ar1=np.arange(1,10).reshape(3,3)
print(ar1)
print('*'*25)
for i in np.nditer(ar1):   # element wise iteration
    print(i)
    print('-'*50)
```

```python
ar1=np.arange(1,10).reshape(3,3)
ar2=ar1.T
print('*'*25)
for i in np.nditer(ar2):   # element wise iteration
    print(i)
    print('-'*50)
```

```python
ar_3d= np.arange(1,25).reshape(2,3,4) # 3d array
ar_3d
```

```python
for i in np.nditer(ar_3d):  # 3d element wise.
    print(i)
```
