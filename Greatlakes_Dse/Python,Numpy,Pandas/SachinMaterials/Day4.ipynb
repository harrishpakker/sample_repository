<span style="font-family: Arial; font-weight:bold;font-size:2.00em;color:#9F1F1B;"> Introduction to  Dataframe

#### Data Frame

* Data is aligned in a tabular fashion in rows and columns.
* DataFrame is a tabular data structure in which data is laid out in rows and column format (similar to a CSV or tables in SQL file), but it can also be used for higher dimensional data sets. 

* The DataFrame object can contain homogenous and heterogenous values, and can be thought of as a logical extension of Series data structures. In contrast to Series, where there is one index, a DataFrame object has one index for column and one index for rows. This allows flexibility in accessing and manipulating data.

#### Renaming the columns and rows using axis

```python
# Syntax:

# dataframe_name.rename({oldname:newname},axis=0 or 1)

# axis=0 => row wise
# axis=1 => columnwise
```

```python
names = ['Ram', 'John', 'Abdul','Siva', 'Adam', 'Rahman']
maths =  [45, 40, 60, 80, 90, 75]
english =[60,75,90,65,80,60]
di={'Names':names,'Maths':maths,'English':english}
scores_df=pd.DataFrame(di)
scores_df
```

```python
scores_df.rename({'Names':'StudentNames'},axis=1)  # axis = 1 (columnwise)
```

```python
scores_df.rename({5:10},axis=0)  # axis = 0 (rowwise)
```

```python
# Tips:

# 1) inplace = True will change the orginal dataframe
# 2) axis = 0 is equivalent to doing with index
# 3) axis = 1 is equivalent to doing with columns
```
