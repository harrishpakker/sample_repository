# Pivot Table

```python
import pandas as pd
```

```python
d = {
    'Name':['Jony','Anil','Sunil','Jony','Anil','Rocky',
            'Jony','Anil','Sunil','Jony','Anil','Rocky'],
    'Exam_Months':['3_Months','3_Months','3_Months','3_Months','3_Months','3_Months',
            '6_Months','6_Months','6_Months','6_Months','6_Months','6_Months'],
     
    'Subject':['Python','Python','SQL','SQL','ML','ML',
               'Python','Python','SQL','SQL','ML','ML'],
   'Score':[62,47,55,74,31,77,85,63,42,67,89,81],
   'Lab':[70,60,63,80,40,77,67,70,80,90,92,95]}
 
df = pd.DataFrame(d,columns=['Name','Exam_Months','Subject','Score','Lab'])
df
```

```python
# Syntax :
# pd.pivotable(dataframe, index =colums, aggfunc='mean')
```

```python
pd.pivot_table(df, index=['Subject'])  # Column
df.info()
df.describe()
df.shape[0]
```

```python
pd.pivot_table(df, index=['Exam_Months'])
```

```python
pd.pivot_table(df, index=['Exam_Months','Subject']) # two columns
```

```python
# Syntax: pd.crosstab(col1,col2)
```

```python
pd.crosstab(df_people['Qualification'], df_people['Hometown'])
```

```python
# Find the crosstab between gender and hometown.
```

```python
# Syntax pd.corsstab(col1,col2,value = numeric colum ,aggfunc = operation)
pd.crosstab(df_people['Hometown'], df_people['Smoker'], values = df_people['Age'], aggfunc = 'mean')
```
