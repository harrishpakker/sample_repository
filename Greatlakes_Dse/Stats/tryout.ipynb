```python
import scipy.stats as stats
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# from scipy.stats import skew 
# from scipy.stats import ttest_rel

before = [40, 35, 38, 37, 33, 30, 28, 35, 35, 40, 33, 35, 36, 36, 40, 38, 35, 30, 30, 38, 39, 35, 
          36, 34, 33, 31, 36, 35, 33, 39]
after = [23, 33, 25, 36, 37, 34, 23, 28, 24, 31, 24, 20, 21, 28, 35, 25, 21, 35, 30, 
         33, 21, 28, 23, 24, 27, 20, 20, 30, 32, 33]

# converting the list to numpy arrays for testing
before_array = np.array(before)
after_array = np.array(after)

# Null hypothesis : Both the average is near meaning that even after speed enforcement activities 
# there is no effect

# Alternate hypothesis : Both the average is very far meaning speed enforcement activities 
# has played an effect reducing the speed

# performing shapiro test for determining if both the data are normally distributed
shapiro(before),shapiro(after)
```

```python
# after data is not normally distributed so wilcoxon is used to infer the data
# if both the data were normally distributed we could use ttest_rel(before_array , after_array)
wilcoxon(before_array - after_array)
```

```python
# Pvalue is less than 0.5(95%confidence),THUS ALTERNATE HYPOTHESIS HOLDS TRUE
# Speed enforcement activities has played a role in reducing the speed
```
