# LA LIGA

```python
laliga = pd.read_csv('Laliga.csv')
```

```python
laliga.columns = laliga.iloc[0]
```

```python
laliga.iloc[0:]
```

```python
laliga = laliga.iloc[1:]
```

```python
laliga.sort_values(by = 'GamesWon',ascending=False)[0:5]
```

```python
laliga[laliga['Debut'].str.contains("193.*",regex = True)]
```

```python
def debut_check(x):
    if (len(x) == 4):
        if (x[:3] == "193" or x[:3] == "194" or x[:3] == "195" or x[:3] == "196" or x[:3] == "197" ):
            return True
        else:
            return False
    else:
        yr = x.split('-')
        if (int(yr[0]) in range(1930,1981) or int(yr[1]) in range(1930,1981)):
            return True
        else:
            return False
```

```python
laliga[laliga['Debut'].apply(debut_check)]
```

```python
laliga['WinningPercentage']= (laliga[laliga['GamesWon'] != "-"]['GamesWon'].astype(int) / 
                     laliga[laliga['GamesPlayed'] != "-"]['GamesPlayed'].astype(int))*100
```

```python
(laliga['GamesWon']== "-").value_counts()
```

```python
(laliga['GamesPlayed'] == "-").value_counts()
```

```python
sns.distplot(laliga['WinningPercentage'])
```

```python
laliga.sort_values(by = 'WinningPercentage',ascending=False)[0:8]
```

```python
samp1_3 = laliga['WinningPercentage'][0:3]

samp4_7 = laliga['WinningPercentage'][3:7]
```

```python
shapiro(samp1_3),shapiro(samp4_7)
```

```python
mannwhitneyu(samp1_3,samp4_7)
# not much difference in winning percentages
```
