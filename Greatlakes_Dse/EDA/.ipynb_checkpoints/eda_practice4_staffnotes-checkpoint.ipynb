```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import Series, DataFrame
```

```python
df1 = sns.load_dataset('tips')
```

```python
df1.head()
```

# Bi-variate and Multi-variate Analysis

### Numerical Numerical Graphs
1. Line
2. Scatter plot

# Line Graph

```python
x1 = [1,2,3,4]
y1 = [5,25,15,36]
```

```python
plt.plot(x1,y1)
plt.show()
```

```python
plt.plot(x1,y1,label='Line 1')
plt.legend()
plt.show()
```

```python
x2 = [1,2,3,4]
y2 = [10,15,36,45]
```

```python
plt.plot(x1,y1,label='Line 1')
plt.plot(x2,y2,label='Line 2')
plt.legend()
plt.show()
```

```python
plt.plot(x1,y1,label='Line 1',color='r',linewidth=5,linestyle='-.',marker='*',markersize=15)
plt.plot(x2,y2,label='Line 2',color='b',linewidth=2,linestyle='-',marker='s',markersize=8)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Simple Line Graph')
plt.legend()
plt.show()
```

# Scatter Plot

```python
df1.head()
```

```python
plt.scatter(df1['total_bill'],df1['tip'])
plt.xlabel('Total Bill')
plt.ylabel('Tips Given')
plt.title('Simple Scatter Graph')
plt.show()
```

```python
plt.scatter(df1['total_bill'],df1['tip'])
plt.axvline(40)
plt.axhline(6)
plt.xlabel('Total Bill')
plt.ylabel('Tips Given')
plt.title('Simple Scatter Graph')
plt.show()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df1)
plt.show()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df1,hue='sex')
plt.show()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df1,hue='day')
plt.show()
```
