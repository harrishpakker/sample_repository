```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df['tb_zscore'] = (df['total_bill'] - df['total_bill'].mean())/df['total_bill'].std()
```

```python
df
```

```python
df[df['tb_zscore'] <-3]
```

```python
df1  = df[df['tb_zscore'] <3]
df1
```

```python
fig,axes = plt.subplots(1,2,figsize = (10,5));
sns.distplot(df['total_bill'],ax = axes[0]);
sns.distplot(df1['total_bill'],ax = axes[1]);
```

```python
# emperical rule
# -1 std to 1 std = 34+34=68% data
# -2 std to 2 std = 47.5+47.5=95% data
# -3 std to 3 std = 49.85+49.85=98.7% data
```
