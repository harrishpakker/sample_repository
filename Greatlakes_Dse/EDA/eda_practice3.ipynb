```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df['sqrt_tb'] = np.sqrt(df['total_bill'])
df['log_tb'] = df['total_bill'].apply(lambda x : np.log(x))
```

```python
df.head()
```

```python
sns.distplot(df['total_bill'])
```

```python
sns.distplot(df['sqrt_tb'])
```

```python
sns.distplot(df['log_tb'])
```

```python
fig,axes = plt.subplots(1,3,figsize = (20,5))
sns.distplot(df['total_bill'],ax = axes[0])
sns.distplot(df['sqrt_tb'],ax = axes[1])
sns.distplot(df['log_tb'],ax = axes[2])
```

```python

tb_skew = df['total_bill'].skew()
tb_sqrt_skew = df['sqrt_tb'].skew()
tb_log_skew = df['log_tb'].skew()


tb_kurt = df['total_bill'].kurt()
tb_sqrt_kurt = df['sqrt_tb'].kurt()
tb_log_kurt = df['log_tb'].kurt()

print(tb_skew,tb_sqrt_skew ,tb_log_skew)
print()
print(tb_kurt,tb_sqrt_kurt,tb_log_kurt)
```
