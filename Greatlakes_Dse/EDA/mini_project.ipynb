```python
df.drop(['Photo','Flag','Club Logo'],axis =1,inplace=True)
```

```python
df['Value']
```

```python
def repl(i):
    if type(i) != float:
    
        if i[-1] == 'M':
            return float(i[1:-1])*1000000
        elif i[-1] == 'K':
            return float(i[1:-1])*1000
```

```python
df['Value'] = df['Value'].apply(repl)
```

```python
df['Value'] = df['Value'].astype(float)
```

```python
df['Wage'] = df['Wage'].apply(repl)
```

```python
df['Wage'] = df['Wage'].astype(float)
```

```python

def date_format(x):
    if x != "" and type(x) != float:
        return x[-4:]
```

```python
# a.sort_values(ascending=False)
df['Joined'] = df['Joined'].apply(date_format)
```

```python
import datetime
def str_to_date(x):
    if type(x) != float:
        return int(x[-4:])
```

```python
df['Contract Valid Until'] = df['Contract Valid Until'].apply(str_to_date)
# df['Contract Valid Until'] = df['Contract Valid Until'].astype(int)
```
