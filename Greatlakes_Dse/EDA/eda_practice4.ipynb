```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip',hue='day',data = df);
```

```python
sns.jointplot(x = 'total_bill', y = 'tip',data = df,kind = 'hex',color = 'red');
```

```python
sns.lmplot(x = 'total_bill', y = 'tip',data = df);
```

```python
sns.lmplot(x = 'total_bill', y = 'tip',data = df,fit_reg=False);
```

```python
sns.lmplot(x = 'total_bill', y = 'tip',data = df,hue='sex');
```

```python
sns.lmplot(x = 'total_bill', y = 'tip',data = df,row='sex');
```

```python
sns.lmplot(x = 'total_bill', y = 'tip',data = df,col= 'smoker',row='sex');
```

```python
sns.lmplot(x = 'total_bill', y = 'tip',data = df,col= 'smoker',row='sex',hue ='day');
```

```python
sns.boxplot(x = 'day', y ='total_bill',data = df)
```

```python
sns.boxplot(x = 'day', y ='total_bill',data = df,hue = 'sex')
```

```python
sns.boxplot(x = 'sex', y ='total_bill',data = df,hue = 'day')
```

```python
sns.violinplot(x = 'day', y ='total_bill',data = df,hue = 'sex')
```

```python
sns.violinplot(x = 'day', y ='total_bill',data = df,hue = 'sex',split=True);
```
