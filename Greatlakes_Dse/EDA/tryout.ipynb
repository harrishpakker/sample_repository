```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
plt.style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = (8,6)
plt.rcParams['font.size'] = 14
plt.rcParams['legend.fontsize'] =  10
```

```python
# importing the drinks csv file which is same as
# the url csv as any files shouldn't be downloaded from external links
df = pd.read_csv('drinks.csv')
```

```python
# displaying first five rows of the dataframe
df.head()
```

```python
# using pivot table to group the data by continent and values contains the fields on which calculations 
# should be performed. IN this case it mean/average as mentioned in the question
df.pivot_table(index = 'continent',values = ['beer_servings','spirit_servings','wine_servings',
                                             'total_litres_of_pure_alcohol'],
               aggfunc='mean' ).plot(kind='bar');
```
