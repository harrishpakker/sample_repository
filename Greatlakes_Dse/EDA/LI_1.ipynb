```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('retail_data.csv',encoding='latin-1')
```

```python
df.head()
```

```python
df['Dealer Name'].value_counts()
```

```python
df_dealer = df[df['Allocation Status'] == 'Dealer']
```

```python
df_dealer['Dealer Name'].value_counts()
```

```python
df_dealer['Dealer Name'].replace(to_replace='pearlson',value='Pearlson',inplace=True)
```

```python
df_dealer['Dealer Name'].value_counts()
```

```python
df_dealer = df_dealer[df_dealer['Dealer Name'] != 'Direct']
```

```python
df_dealer['Dealer Name'].value_counts()
```

```python
df_dealer
```

```python
pd.crosstab(df_dealer['Dealer Name'],df_dealer['Status'])
```

```python
pd.crosstab(df_dealer['Dealer Name'],df_dealer['Status']).plot(kind = 'bar');
```

```python
conversion_rate = pd.crosstab(df_dealer['Dealer Name'],df_dealer['Status']).apply(lambda x : x*100/x.sum(),axis = 1)
```

```python
conversion_rate
```
